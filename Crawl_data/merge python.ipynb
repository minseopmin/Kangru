```python
import pandas as pd

# 카카오 데이터 불러오기
kakao_data = pd.read_csv('Kakao_data.csv')

# 구글 데이터 불러오기
google_data = pd.read_csv('Google_data.csv')

# 카카오 데이터와 구글 데이터를 합치기
merged_data = pd.merge(kakao_data, google_data, on=['Store', 'keyword_food', 'keyword_loc'], how='inner')

# 카카오 데이터에서 phone_num과 address 가져와서 데이터 프레임에 추가하기
phone_num_list = []
address_list = []
for i in merged_data.index:
    store_name = merged_data.at[i, 'Store']
    keyword_food = merged_data.at[i, 'keyword_food']
    keyword_loc = merged_data.at[i, 'keyword_loc']
    temp = kakao_data.loc[(kakao_data['Store'] == store_name) & 
                          (kakao_data['keyword_food'] == keyword_food) & 
                          (kakao_data['keyword_loc'] == keyword_loc)]
    if not temp.empty:
        phone_num_list.append(temp['phone_num'].values[0])
        address_list.append(temp['address'].values[0])
    else:
        phone_num_list.append('')
        address_list.append('')

merged_data['phone_num'] = phone_num_list
merged_data['address'] = address_list

# 필요한 컬럼만 선택
merged_data = merged_data[['Store', 'keyword_food', 'keyword_loc', 'score_x', 'comment_x', 'score_y', 'comment_y', 'phone_num', 'address']]

# 컬럼 이름 변경
merged_data = merged_data.rename(columns={'score_x': 'kakao_score', 'comment_x': 'kakao_comment', 
                                          'score_y': 'google_score', 'comment_y': 'google_comment'})

merged_data.to_csv('Merge_data_store.csv',index=True)
```
